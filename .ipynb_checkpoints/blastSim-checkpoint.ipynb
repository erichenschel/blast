```python
# R = S(gamma) * E^(1/5) * (rho_o)^(-1/5)*t^(2/5)
# initial conditions::
# u(0, r) = 0 m/s
# p(0, r) = 101 kPa
# rho(0, r) = 1.225 kg/m^3


# E = 5.36*rho_o*A^2
```

```python
import matplotlib.pyplot as plt
```

```python
# specific heat of dry air
gamma = 1.4 
u_o = 0 # m/s
p_o = 101. # kPa
rho_o = 1.225 # kg/m^3
```

```python
# list of nuclear weapon energy outputs
# sourced: https://en.wikipedia.org/wiki/Nuclear_weapon_yield
# included in list in order {Davy Crockett, AIR-2 Genie, 
# Hiroshima("Little Boy"), Nagasaki("Fat Man"), W76 warhead,
# W87 warhead, W88 warhead, Ivy King (fission bomb), Orange Herald Small,
# B83 nuke, B53 nuke, Tsar Bomba, Total energy expelled 
# during all nuke testing}

atoms = [
    0.02,
    1.5,
    18.0,
    22.0,
    100.0,
    300.0,
    475.0,
    500.0,
    720.0,
    1200.0,
    9000.0,
    50000.0,
    510300.0]
```

```python
# Energy Kilotons --> Joules
def energyConversion(E):
    return E * (4.184*10**12)
```

```python
# radius of shock wave
def radius(t, energy):
    rho_o = 1.225 # kg/m^3
    return (1.031) * energy**(1./5) * rho_o**(-1./5) * t**(2./5)
```

```python
# B is taken from 
# https://www3.nd.edu/~powers/ame.90931/taylor.blast.wave.I.pdf
# pg. 167
def velocity(radius, energy):
    rho_o = 1.225 # kg/m^3
    B = 2.31 # 0.926 * (5/2)
    return radius**(-3./2) * energy**(1/2) * (B*rho_o)**(-1./2)
```

```python
# max pressure
def pressure(radius, energy):
    return 0.155*radius**(-3.)*energy
```

```python
# simulating radial expansion w.r.t. time
# velocity w.r.t. radius
# pressure w.r.t. radius
for out in atoms:
    print(out)
    E = energyConversion(out)
    rad = []
    vel = []
    pres = []
    for i in range(100):
        t = float(i)/(10**7)

        r = radius(t, E)
        rad.append(r)
    for i in range(1, 20):
        r = float(i)/(10**7)

        v = velocity(r, E)
        vel.append(v)

        p = pressure(r, E)
        pres.append(p)

    # pot radius w.r.t. time
    plt.scatter(range(len(rad)), rad)
    plt.show()
    # plot velocity w.r.t. radius
    plt.scatter(range(len(vel)), vel)
    plt.show()
    # plot pressure w.r.t. radius
    plt.scatter(range(len(pres)), pres)
    plt.show()
```
